### Exercise 4

Next, iterate through the list `counties_list`, print out each dictionary if the `name` value is "King County," and use your `report_evictions_info()` function to report the number of evictions in the county. Note: you will need to work with the dictionary keys `name`, `evictions`, and `year`.

I'm including the function here just to remember what it does:

```python
def report_evictions_info(num_evictions, county, year):
    print(f"There were {num_evictions} evictions in {county} in {year}.")
```

```python
for county in counties_list:
    if county['name'] == 'King County':
        
        # Get the name of the county by referencing the key "name"
        county_name = county['name']
        # Get the number of evictions by referencing the key "evictions"
        evictions = county['evictions']
        # Get the year by referencing the key "year"
        year = county['year']
        
        report_evictions_info(evictions, county_name, year)
```

### Exercise 5

Next, iterate through the list `counties_list`, print out each dictionary if the `name` value is "King County," and then use your `calculate_eviction_rate()` function to return the eviction rates. Note: you will need to work with the dictionary keys `evictions` and `renter.occupied.households`.

I'm including the function here just to remember what it does:

```python
def calculate_eviction_rate(num_evictions, renting_households):
    return num_evictions / renting_households
```

```python
for county in counties_list:
    if county['name'] == 'King County':
        
        # Get the number of evictions by referencing the key "evictions"
        evictions = county['evictions']
         # Get the number of households by referencing the key "renter.occupied.households"
        households = county['renter.occupied.households']
        
        eviction_rate = calculate_eviction_rate(evictions, households)
        
        print(eviction_rate)
```
